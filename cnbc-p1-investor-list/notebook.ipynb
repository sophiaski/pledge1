```python
import pandas as pd 
```

```python
df = pd.read_csv('p1_cnbc_top_50.csv')
df = df[['cnbc_link', 'company', 'Investors']]
Investors = df.Investors
```

```python
# Create long list of VCs, remove duplicates
string=''
for i in range(len(Investors)):
    string += '; '+str(Investors[i])
new_cols = sorted(list(set(string.split('; '))))[1:]
```

```python
# Create a row for each unique VC value
df=pd.concat([df,pd.DataFrame(columns=new_cols)]).fillna(0)
```

```python
# Populate the empty VC rows
for i in range(len(Investors)):
    a_list = Investors[i].split('; ')
    for item in a_list:
        df[item][i] = 1
```

```python
# Create count matrix
matrix = df.loc[:,new_cols]
matrix.loc['Total',:]= matrix.sum(axis=0)
matrix.loc[:,'Total'] = matrix.sum(axis=1)
matrix
```

```python
sum_counts = list(matrix.loc['Total',:])[:-1] #Remove total cell
sum_counts[sum_counts.index(max(sum_counts))]
print("What VC invested in the CNBC list the most?",new_cols[138])
```

```python
# Create new dataframe to filter by total counts
df_sum = pd.DataFrame(dict(zip(new_cols,sum_counts)),index=[0])
df_sum = df_sum[df_sum > 2].dropna(axis=1)
```

```python
# Print out of top VCs with more than 2 hits on the CNBC list
df_sum.T[0].astype(int)
```

```python
top_VC=df_sum.T[0].index.to_list()
top_cols=['company']+top_VC
df[top_cols]
```

```python
df[top_cols].to_csv('top_VC_CNBC_50_list.csv')
```
