### Describe available actions from Ticker datatype

The number of shares actually available to trade is known as float. There are also restricted shares, which are set aside for employee compensation and incentives. Restricted shares are also part of authorized shares. The total number of a company's outstanding shares as seen in the balance sheet is the sum of float and restricted shares.

```python
# Get the stock data of P1 companies that went through an IPO
for company in companies.keys():
    try:
        stock = yf.Ticker(companies[company]['tick'])
        stock_hist = yf.download(tickers=companies[company]['tick'], start=companies[company]['ipo'])
    except KeyError:
        print(f'{company} has not had a liquidity event. Skipping plot!')
        continue    
    stock_info = {key: stock.info.get(key) for key in ['floatShares', 'sharesOutstanding', 'marketCap']}
    stock_info['restrictedShares'] = stock_info['sharesOutstanding'] - stock_info['floatShares']
    print('Data for {}'.format(company))
    print(companies[company]['ipo_dt'].strftime('IPO Date: %b %d, %Y'))
    #print('Market Cap (Today): ${:,.0f}'.format(stock_info['marketCap'])) # This is off for some reason
    print('Market Cap (Today): ${:,.0f}'.format(stock_hist.Close[-1]*stock_info['sharesOutstanding']))
    print('Outstanding Shares: {:,.0f}'.format(stock_info['sharesOutstanding']))
    print('Share price (Today): ${:,.2f}'.format(stock_hist.Close[-1]))
    print('Share price (At IPO): ${:,.2f}'.format(stock_hist.Close[0]))
    #try:
    #    print('Social Impact Shares: {:,.0f}'.format(companies[company]['p1_shares']))
    #    print('Social Impact Value (Today): ${:,.0f}'.format(companies[company]['p1_shares']*stock_hist.Close[-1]))
    #    print('Social Impact Value (At IPO): ${:,.0f}'.format(companies[company]['p1_shares']*stock_hist.Close[0]))
    #except KeyError:
    #    print('We do not know the number of social impact shares of {}.\nEstimating using 1% of outstanding shares now and at IPO.'.format(company))
    #    companies[company]['est_p1_shares'] = stock_info['sharesOutstanding']*0.01
    #    print('Estimated Social Impact Shares: {:,.0f}'.format(companies[company]['est_p1_shares']))
    #    print('Estimated Social Impact Value (Today): ${:,.0f}'.format(companies[company]['est_p1_shares']*stock_hist.Close[-1]))
    #    print('Estimated Social Impact Value (At IPO): ${:,.0f}'.format(companies[company]['est_p1_shares']*stock_hist.Close[0]))
    stock_hist.Close.plot()
    plt.show()
```
