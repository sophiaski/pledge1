```python
import pandas as pd 
import re
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('p1_cnbc_top_50.csv')
df = df[['cnbc_link', 'company', 'Investors']]
Investors = df.Investors
```

```python
# Create long list of VCs, remove duplicates
string=''
for i in range(len(Investors)):
    string += '; '+str(Investors[i])
new_cols = sorted(list(set(string.split('; '))))[1:]
```

```python
# Create a row for each unique VC value
df=pd.concat([df,pd.DataFrame(columns=new_cols)]).fillna(0)
```

```python
# Populate the empty VC rows
for i in range(len(Investors)):
    a_list = Investors[i].split('; ')
    for item in a_list:
        df[item][i] = 1
```

```python
# Create count matrix
vals = df.loc[:,new_cols]
vals.loc['Total',:]= vals.sum(axis=0)
vals.loc[1:,'Total'] = vals.sum(axis=1)
vals
```
