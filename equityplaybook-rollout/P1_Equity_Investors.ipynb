```python
import pandas as pd 
```

```python
df = pd.read_csv('data.csv')
df = df[['Account Name', 'Investors']]
df=df.drop([2,6,14,15,16,22,23,24,26,28,29,30,31,32,33,37,38,39,40], axis=0)
df.reset_index(inplace=True, drop=True)
Investors = df.Investors

# Create long list of VCs, remove duplicates
string=''
for item in Investors:
    string += '; '+str(item)
Unique_Investor_List = sorted(list(set(string.split('; '))))[1:]

Unique_Investor_List[:5]
```

```python
# Create a row for each unique VC value
df=pd.concat([df,pd.DataFrame(columns=Unique_Investor_List)]).fillna(0)
df.head()
```

```python
# Populate the empty VC rows
for i in range(len(Investors)): # By each company
    a_list = Investors[i].split('; ') # By each VC
    for item in a_list:
        df[item][i] = 1
df.head()
```

```python
# Create count matrix
df_matrix = df.loc[:,Unique_Investor_List]
df_matrix.loc['Total',:]= df_matrix.sum(axis=0)
df_matrix.loc[:,'Total'] = df_matrix.sum(axis=1)
# Add back company names
df_matrix=df_matrix.reindex(columns=list(['Account Name']+ df_matrix.columns.tolist()))
df_matrix['Account Name'] = df['Account Name']
df_matrix.iloc[28,0]='All'
df_matrix.tail()
```

```python
sum_counts = list(df_matrix.loc['Total',:])[1:-1] #Pull out 2 total cell
max_index = sum_counts.index(max(sum_counts))
print("What VC invested in the P1 Equity Pledge Company list the most?",Unique_Investor_List[max_index])
```

```python
# Create new dataframe to filter by total counts
df_sum = pd.DataFrame(dict(zip(Unique_Investor_List,sum_counts)),index=[0])
df_sum = df_sum[df_sum > 2].dropna(axis=1)
# Print out of top VCs with more than 2 hits on the CNBC list
print(df_sum.T[0].astype(int))
# Save top VC list
df_top_VC = df_sum.T
```

```python
# Create smaller dataframe with just the top VCs
top_VC=df_sum.T[0].index.to_list()
top_cols=['Account Name']+top_VC
df_matrix_top = df[top_cols]
# Add Total columns
df_matrix_top.loc['Total',:]= df_matrix_top.sum(axis=0)
df_matrix_top.loc[:,'Total'] = df_matrix_top.sum(axis=1)
df_matrix_top.iloc[28,0]='All'
df_matrix_top
```

```python
df_top_VC.to_csv("top_VC_list.csv") # Top hit list of VCs
df_matrix_top.to_csv('top_company_matrix.csv') # Top VC company matrix
df_matrix.to_csv('company_matrix.csv') # Company matrix
```
