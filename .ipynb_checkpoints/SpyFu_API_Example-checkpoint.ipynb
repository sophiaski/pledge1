```python
df = pd.read_csv('All_SEM_Data_8_5_2020.csv')               # Update to current report csv (sometimes encoding = "ISO-8859-1")
df['Domain Name'] = df['Domain Name'].str.replace('https://', '')
df['Domain Name'] = df['Domain Name'].str.replace('http://', '')
df.head()
```

```python
df_smol = df[['Domain ID', 'Account ID', 'Domain Name', 'Account Name']].copy()
```

```python
df_smol.tail(10)
```

```python
df_smol.head()
```

```python
df_smol2 = df_smol[6699:]
df_smol2.head()
```

```python
rows = df_smol2.T.to_dict().values()

counter = 1
for row in rows:
    print('%d - Getting info for %s' % (counter, row['Account Name']))
    counter += 1
    
    row['SpyFu_Historical_Date_Run__c'] = str(dt.datetime.today())[:10]
    
    r = requests.get(url = URL, params=dict(
        domain = row['Domain Name'],
        api_key = Key,
        months = 13
    ))
    if r.status_code == 200:
        result_dict = json.loads(r.content)
        
        # Update the month to 1 yr ago before running
        if len(result_dict) == 13 and result_dict[0]['date'] == "Aug '19": 
            row['Spyfu_monthly_adwords_budget__c'] = result_dict[-1]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_1mo__c'] = result_dict[-2]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_2mo__c'] = result_dict[-3]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_3mo__c'] = result_dict[-4]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_4mo__c'] = result_dict[-5]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_5mo__c'] = result_dict[-6]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_6mo__c'] = result_dict[-7]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_7mo__c'] = result_dict[-8]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_8mo__c'] = result_dict[-9]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_9mo__c'] = result_dict[-10]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_10mo__c'] = result_dict[-11]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_11mo__c'] = result_dict[-12]['estimatedMonthlyBudget']
            row['Spyfu_adwords_budget_12mo__c'] = result_dict[-13]['estimatedMonthlyBudget']
        elif len(result_dict) > 0:
            row['Spyfu_monthly_adwords_budget__c'] = result_dict[-1]['estimatedMonthlyBudget']
            
        if counter % 100 == 0:
            print("Saving to DataFrame")
            df_out = pd.DataFrame(list(rows))
            print('df shape:', df_out.shape)
            print(df_out.columns)
            print(df_out.isnull().sum())

df_out = pd.DataFrame(list(rows))   
```
